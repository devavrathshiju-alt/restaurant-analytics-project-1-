```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("zomato.csv")
```

```python
df.head()
df.info()
```

```python
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
```

```python
df.columns = [
    "restaurant_name",
    "restaurant_type",
    "rating",
    "num_of_ratings",
    "avg_cost_for_two",
    "online_order",
    "table_booking",
    "cuisines",
    "area",
    "local_address"
]
```

```python
display(df.head())
```

```python
df.isnull().sum()
```

```python

df["rating"] = df["rating"].fillna(df["rating"].median())
df["num_of_ratings"] = df["num_of_ratings"].fillna(0)
df["avg_cost_for_two"] = df["avg_cost_for_two"].fillna(df["avg_cost_for_two"].median())

for col in ["restaurant_type", "cuisines", "area", "local_address"]:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df["rating"] = df["rating"].astype(float)
df["num_of_ratings"] = df["num_of_ratings"].astype(int)
df["avg_cost_for_two"] = df["avg_cost_for_two"].astype(int)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df["rating_category"] = np.where(
    df["rating"] >= 4.0, "High Rated",
    np.where(df["rating"] >= 3.0, "Average Rated", "Low Rated")
)
```

```python
df.head()
df.info()
```

```python
df.to_csv("cleaned_dataset.csv", index=False)
```
